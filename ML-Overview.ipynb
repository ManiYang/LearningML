# ML Introduction
- Machine Learning is the science and art of programming computers so that they can learn from data.
- Machine learning *is the field of study that gives computers the ability to learn without explicitly programmed.*  — Arthur Samuel, 1959
- *A computer program is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P, improves with experience E.* — Tom Mitchell, 1997

Traditional Approach
- need to figure out and implement a long list of complicated rules, which is hard to maintain
- keep writing new rules to adapt to changes
- problem can be too complicated (e.g., speech recognition)

Machine Learning
- shorter program, easier to maintain
- adapting to change with new data
- learn complicated problems by itself
- can help us learn (reveal unsuspected correlations or new trends)

# Types of ML Systems
- with human supervision?
  - supervised learning
  - unsupervised learning
  - semisupervised learning
  - Reinforcement Learning
- can learn incrementally on the fly?
  - batch learning
  - online learning
- simply compare new data to known data, or build predictive model?
  - instance-based learning
  - model-based learning

## Supervised/Unsupervised Learning

### Supervised Learning

In supervised learning, the training data includes the desired solutions, called *labels*.

Typical tasks:

- classification
- regression
  - *features, attributes, predictors*

Some supervised learning algorithms:

- k-Nearest Neighbors
- Linear Regression
- Logistic Regression
- Support Vector Machines (SVMs)
- Decision Trees & Random Forests
- Neural Networks (but not all neural network architectures are supervised)

### Unsupervised Learning

In unsupervised learning, the training data is unlabeled.

Some unsupervised learning algorithms:

- Clustering
  - k-Means
  - Hierarchical Cluster Analysis (HCA)
  - Expectation Maximization
- Visualization and dimensionality reduction
  - Principal Component Analysis (PCA)
  - Kernel PCA
  - Locally-Linear Embedding (LLE)
  - t-distributed Stochastic Neighbor Embedding (t-SNE)
- Association rule learning
  - Apriori
  - Eclat

Typical tasks:

- Visualization
  - complex unlabeled data  →  2D or 3D representation
  - preserving as much structure as possible.
- Dimensionality reduction
  - simply the data without losing too much information
  - feature extraction
  - It is often a good idea to try to reduce dimensionality of training data before we feed them to another ML algorithm.
- Anomaly detection
- Association rule learning

### Semisupervised Learning

Most semisupervised learning algorithms are combinations of unsupervised and supervised algorithms.

### Reinforcement Learning

An *agent* observe the environment, select and perform actions, and get *rewards* or *panelties* in return. It must learn by itself what is the best strategy (what action to choose in a given situation), called *policy*, to get the most reward over time.

## Batch and Online Learning

### Batch Learning

- trained using all the available data  →  launched, running without learning anymore
- new data  →  train a new version of the system  →  replace the old running program
- may need a lot of computing resources

If the system need to adapt to rapidly changing data or has limited resources, then online learning is more suitable.

### Online Learning
Train the system incrementally by feeding it data instances sequentially (either individually or by small groups called *mini-batches*).

Online learning can be used when

- the system need to receive data as a continuous flow and adapt change rapidly or autonomously
- the computing resources are limited 
- training a huge data set that cannot fit in one machine’s main memory (*out-of-core* learning). (This is usually done offline, i.e., not on the live system. “Online learning” can be a confusing name.)

Learning rate

- high learning rate  →  quickly adapt to new data & forget old data
- low learning rate  →  have more inertia, less sensitive to noise or non representative data points

If bad data is fed to the system, the system’s performance will gradually decline. We may want to monitor the system’s performance and the input data.

## Instance-Based/Model-Based Learning
Generalization: from performing well on training data to performing well on new instances.

### Instance-Based Learning
The system learns the training examples by heart, then generalizes to a new case by finding which training examples it is similarity to.

### Model-Based Learning
Build a model from training examples, then use this model to make predictions.

# Main Challenges of ML
“bad algorithm” & “bad data”

## Insufficient Quantity of Training Data
Even for very simple problems we typically need thousands of examples, and for complex problems (such as image or speech recognition) we may need millions of examples.

## Nonrepresentative Training Data
- It is crucial to use a training set that is representative of the cases you want to generalize to.
- Beware of *sampling bias*.

## Poor-Quality Data
Training data may contain errors, outliers, noise, and missing data.

## Irrelevant Features
Our system will only be capable of learning if the training data contains enough relevant features and not too many irrelevant ones.

Feature engineering:

- feature selection
- feature extraction: combining existing features to produce a more useful one (dimensionality reduction algorithms can help)
- creating new feature by gathering new data

## Overfitting the Training Data
Possible solutions:

- simplify the model
  - fewer parameters
  - reduce number of features
  - constrain the model (regularization)
- gather more training data
- reduce noise in the training data

## Underfitting the Training Data
Possible solutions:

- select a more powerful model
- feed better features to the learning algorithm
- reducing the constrains no the model

# Testing and Validation
- Split the data into the *training set* and *test set* (e.g., 80%, 20%), and use the test set to estimate the generalization error rate.
- If the training error is low but the generalization error is high, it means that the model is overfitting the training data.
- *Validation set*
- *Cross-validation*: The training set is split into complementary subsets, and each model is trained against a combination of these subsets and validated against the remaining subsets. Once the model type and hyperparameters have been selected, a final model is trained using these hyperparameters on the full training set, and the generalized error is measured on the test set.
